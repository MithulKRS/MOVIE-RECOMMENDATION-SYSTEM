```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("tmdb_5000_movies.csv")
c_data=pd.read_csv("tmdb_5000_credits.csv")
```

```python
data.head()
```

```python
c_data.head()
```

```python
data=data.merge(c_data,on='title')
```

```python
data.head(1)
```

```python
#genre
#id
#keywords
#title
#overview
#cast
#crew
data=data[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.iloc[0].genres
```

```python
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
data['genres']=data['genres'].apply(convert)
```

```python
data
```

```python
data['keywords']=data['keywords'].apply(convert)
```

```python
import ast
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
data['cast']=data['cast'].apply(convert3)
```

```python
data.head()
```

```python
import ast
def fetch(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
    return l
```

```python
data['crew']=data['crew'].apply(fetch)
```

```python
data.head()
```

```python
data['overview'].apply(lambda x:x.split())
```

```python
data['genres']= data['genres'].apply(lambda x:[i.replace(" ","") for i in x])
data['cast']= data['cast'].apply(lambda x:[i.replace(" ","") for i in x])
data['crew']= data['crew'].apply(lambda x:[i.replace(" ","") for i in x])
data['keywords']= data['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data['overview'] = data['overview'].apply(lambda x:x.split())
```

```python
data
```

```python
data['tags'] = data['overview'] + data['genres'] + data['keywords'] + data['cast'] + data['crew']
```

```python
data
```

```python
new = data.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags']=new['tags'].apply(lambda x:" ".join(x))
```

```python
new
```

```python
new['tags']=new['tags'].apply(lambda x:x.lower())
```

```python
new
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem(text):
    y=[]
    for i in  text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tags']=new['tags'].apply(stem)
```

```python
new['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('John Carter')
```
